```python
# correlation

import plotly.graph_objects as go
import cufflinks as cf
import pandas as pd
cf.go_offline(connected=True)

df = pd.read_csv('../documents/00_Material(Uploaded)/COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/04-01-2020.csv', encoding='utf-8-sig')
print(df.head())

# 1) heatmap
df_corr = df.corr()
df_corr.iplot(kind='heatmap')

# cf.help('heatmap')
cf.colors.scales()
df_corr.iplot(kind='heatmap', colorscale='ylorrd')

import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(
    go.Heatmap(    
        x=df_corr.index,
        y=df_corr.columns,
        z=df_corr,
        colorscale='reds'
    )
)

fig.show()

# 2) scatter
df.iplot(kind='scatter', x='Recovered', y='Confirmed', mode='markers')

# cf.help('scatter')

import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(
    go.Scatter(    
        x=df['Recovered'],
        y=df['Confirmed'],
        mode='markers'
    )
)

fig.show()
```
